```python
import requests
from io import BytesIO
import cairosvg
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from PIL import Image
from google.colab import drive
drive.mount('/content/drive')
```

```python
def download_svg(url):
    response = requests.get(url)
    svg_content = response.content
    image = Image.open(BytesIO(cairosvg.svg2png(bytestring=svg_content)))
    return image

def download_svg_from_path(file_path):
    with open(file_path, 'rb') as file:
        svg_content = file.read()
    image = Image.open(BytesIO(cairosvg.svg2png(bytestring=svg_content)))
    return image

def combine_images(images):
    widths, heights = zip(*(i.size for i in images))
    combined_width = max(widths)
    combined_height = sum(heights)
    combined_image = Image.new('RGB', (combined_width, combined_height), 'white')

    y_offset = 0
    for image in images:
        combined_image.paste(image, (0, y_offset))
        y_offset += image.size[1]

    return combined_image

def Combined_Visualization(titles):
  for title in titles:
    image_list = []
    metrics= ["accuracy","auc","f1_score","precision","recall"]
    for metric in metrics:
      image_path = "/content/drive/My Drive/Colab Notebooks/Muawiya/Genetic Programming Combiner with DFS/Images/{}/{}_{}.svg".format(title[1],title[0], metric)
      image = download_svg_from_path(image_path)
      image_list.append(image)
    combined_image = combine_images(image_list)
    final_image_path = "/content/drive/My Drive/Colab Notebooks/Muawiya/Genetic Programming Combiner with DFS/Images/{}/{}_{}.png".format(title[1],title[0], "combined_image")
    combined_image.save(final_image_path)
```

```python
titles=[["Gradual Drift With Evolving","Gradual Drift"],
        ["Gradual Drift Without Evolving","Gradual Drift"],
        ["Blip Drift With Evolving","Blip Drift"],
        ["Blip Drift Without Evolving","Blip Drift"],
        ["Recurrent Drift With Evolving","Recurrent Drift"],
        ["Recurrent Drift Without Evolving","Recurrent Drift"],
        ["Sudden Drift With Evolving","Sudden Drift"],
        ["Sudden Drift Without Evolving","Sudden Drift"],
        ["Incremental Drift With Evolving","Incremental Drift"],
        ["Incremental Drift Without Evolving","Incremental Drift"],
        ["kddcup99 Drift With Evolving","kddcup99"],
        ["kddcup99 Drift Without Evolving","kddcup99"],
        ["ISCX2012 With Evolving","ISCX2012"],
        ["ISCX2012 Without Evolving","ISCX2012"],
        ["CSE_CIC2018 With Evolving","CSE_CIC2018"],
        ["CSE_CIC2018 Without Evolving","CSE_CIC2018"],
        ["CICIDS2017 With Evolving","CICIDS2017"],
        ["CICIDS2017 Without Evolving","CICIDS2017"],
        ]
Combined_Visualization(titles)
```
